```python
# Importing necessary libraries
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1: Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5.
# Parameters
mean = 50
std_dev = 5
confidence_level = 0.95

# Calculate margin of error
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * std_dev

# Confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print("Q1: 95% Confidence Interval:", confidence_interval)
```

```python
# Q2: Conduct a chi-square goodness of fit test for M&Ms color distribution.
# Observed and expected frequencies
observed = [22, 18, 21, 12, 9, 18]  # Example observed data
expected = [20, 20, 20, 10, 10, 20]  # Expected proportions

# Chi-square test
chi_square_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
print("\nQ2: Chi-Square Statistic:", chi_square_stat, ", p-value:", p_value)
```

```python
# Q3: Calculate chi-square statistic and p-value for a contingency table.
# Contingency table
data = np.array([[20, 15], [10, 25], [15, 20]])

# Chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(data)
print("\nQ3: Chi-Square Statistic:", chi2_stat, ", p-value:", p_val)
```

```python
# Q5: Calculate the 90% confidence interval for a sample with a mean of 75 and a standard deviation of 12.
# Parameters
mean = 75
std_dev = 12
confidence_level_90 = 0.90

# Calculate margin of error
z_score_90 = stats.norm.ppf(1 - (1 - confidence_level_90) / 2)
margin_of_error_90 = z_score_90 * std_dev

# Confidence interval
confidence_interval_90 = (mean - margin_of_error_90, mean + margin_of_error_90)
print("\nQ5: 90% Confidence Interval:", confidence_interval_90)
```

```python
# Q6: Plot the chi-square distribution with 10 degrees of freedom.
# Parameters
df = 10
x = np.linspace(0, 30, 500)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, stats.chi2.pdf(x, df), label="Chi-Square Distribution (df=10)")
plt.fill_between(x, stats.chi2.pdf(x, df), where=(x <= 15), alpha=0.5, color='red', label='Shaded Area (Chi-square <= 15)')
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
