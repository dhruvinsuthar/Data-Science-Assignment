```python
# Q1. What is the Probability Density Function (PDF)?
# The Probability Density Function (PDF) describes the likelihood of a continuous random variable taking a specific value. 
# It is a function that gives the relative likelihood for this variable to take on a given value. 
# For continuous variables, the area under the PDF curve between two values represents the probability of the variable falling within that range.
```

```python
# Q2. What are the types of Probability Distribution?
# The main types of probability distributions are:
# - Discrete distributions: e.g., Binomial, Poisson, Geometric
# - Continuous distributions: e.g., Normal, Exponential, Uniform
# - Multivariate distributions: e.g., Multinomial, Multivariate normal
```

```python
# Q3. Python function to calculate the PDF of Normal Distribution:
import math

def normal_pdf(x, mean, std_dev):
    return (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
```

```python
# Q4. Properties of Binomial Distribution:
# Properties:
# - Fixed number of trials (n)
# - Each trial has two possible outcomes (success/failure)
# - Probability of success is constant (p)
# - Trials are independent
# Examples:
# 1. Flipping a coin 10 times and counting the number of heads.
# 2. Testing a drugâ€™s effectiveness on 100 patients.
```

```python
# Q5. Random sample from Binomial distribution and histogram:
import numpy as np
import matplotlib.pyplot as plt

data = np.random.binomial(n=1, p=0.4, size=1000)
plt.hist(data, bins=2, edgecolor='black')
plt.title('Binomial Distribution Histogram (p=0.4)')
plt.xlabel('Success (1) or Failure (0)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q6. Python function to calculate the CDF of Poisson Distribution:
from scipy.stats import poisson

def poisson_cdf(k, mean):
    return poisson.cdf(k, mean)
```

```python
# Q7. Difference between Binomial and Poisson Distribution:
# - Binomial distribution models the number of successes in a fixed number of trials.
# - Poisson distribution models the number of events occurring in a fixed interval of time or space, given a constant rate of occurrence.
# - Poisson distribution can be considered as an approximation of Binomial when the number of trials is large and the probability of success is small.
```

```python

# Q8. Random sample from Poisson distribution and calculate mean and variance:
poisson_data = np.random.poisson(lam=5, size=1000)
mean = np.mean(poisson_data)
variance = np.var(poisson_data)

print(f"Sample Mean: {mean}")
print(f"Sample Variance: {variance}")
```
