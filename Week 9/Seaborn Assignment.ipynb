Line Plot: Used to represent data trends or changes over time by showing the relationship between two continuous variables.

Bar Plot: Used to display the mean or aggregated values of a quantitative variable across different categories.

Box Plot: Helps visualize the distribution of data, including its median, quartiles, and potential outliers.

Histogram: Used to display the frequency distribution of a single continuous variable.

Heatmap: A graphical representation of data where individual values are represented by variations in color intensity.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Line Plot")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Box Plot: Pclass vs Age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot: Pclass vs Fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.tight_layout()
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut", kde=False, palette="viridis")
plt.title("Price Distribution of Diamonds by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species", palette="coolwarm")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Flights Dataset")
plt.show()
```
