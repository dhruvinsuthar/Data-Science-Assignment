Matplotlib is a popular Python library for data visualization. It provides tools for creating a wide variety of plots and charts, making it easier to understand trends and relationships in data. it uses to visualize data in the form of graphs and plots, to make insights from data more comprehensible, to customize plots with labels, titles, legends, etc., five plots that can be created using the Pyplot module.

Five plots using Pyplot module are:

Line Plot

Bar Plot

Scatter Plot

Histogram

Box Plot

A scatter plot is a type of plot that displays data points as individual dots on a two-dimensional plane. It is used to visualize relationships or correlations between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, color='blue', label="Data points")

plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

The subplot() function is used to create multiple plots in a single figure. It allows you to divide the figure into a grid of subplots, enabling you to display multiple visualizations side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1, color="blue")
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2, color="green")
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3, color="red")
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4, color="purple")
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

A bar plot is used to represent categorical data with rectangular bars. The length or height of the bar is proportional to the value it represents. It is used to compare values across categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit, color='orange')
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.barh(company, profit, color='blue')
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

A box plot is a graphical representation of data distribution that shows the minimum, first quartile, median, third quartile, and maximum. It is used to identify outliers and visualize the spread and skewness of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
